### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = x ** 2

# Input validation
if not (isinstance(x, (list, np.ndarray)) and isinstance(y, (list, np.ndarray))):
    raise TypeError("x and y must be lists or NumPy arrays.")
if len(x) != len(y):
    raise ValueError("x and y must have the same length.")

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x²', color='blue')

# Customizations
plt.title('Line Plot of y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.show()
```
