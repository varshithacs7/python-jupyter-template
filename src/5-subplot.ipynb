### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for all plots
x_line = np.linspace(0, 10, 100)
y_line = x_line ** 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Input validation for scatter plot data
if len(x_scatter) != len(y_scatter):
    raise ValueError("Scatter plot x and y values must have the same length.")

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x_line, y_line, color='blue')
axs[0, 0].set_title('Line Plot')

# Bar chart
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Ensure pie is circular

# Layout adjustment
plt.tight_layout()
plt.show()
```
